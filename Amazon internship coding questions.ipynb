## This tree take the list of nodes in level traversal and then return level at which maximum sum of nodes be present

```python
def maximum_level_sum(l):
    n=1 #used for level counting
    s=l[0]
    level=0
    c=1 # used for counting maximum number of node present in that particular levl
    i=1 # used for iterating over an queue generated by level order traversal
    while True:
        c=2**n
        if(c<len(l)):
            if(s<sum(l[i:i+c])):
                level=n
            s=max(s,sum(l[i:i+c]))
        else:
            if(s<sum(l[i:i+c])):
                level=n
            s=max(s,sum(l[i:i+c]))
            break
        n=n+1
        i+=c
    return s,level
```

## Fibonacci with recursion

```python
def fibonacci(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)
print(fibonacci(10))
```

## Stock price maximum profit problem

```python
ans=[]
def calculate(l):
    while True:
        max_index=l.index(max(l))
        min_index=l.index(min(l))
        if(max_index>min_index):
            return (l[max_index]-l[min_index])
        if(max_index==min_index):
            return 0
        else:
            l1=l[:max_index+1]
            l2=l[max_index+1:]
            if l1:
                val=calculate(l1)
                if val:
                    ans.append(val)
            if l2:
                val1=calculate(l2)
                if val1:
                    ans.append(val1)
        break
l=[100,200,250,90,45,96,28]
calculate(l)
print(max(ans))
```

## Left view of a binary tree

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    def left_view(self,start):
        print("Left view of a tree")
        while start:
            print(start.data,end=' ')
            start=start.left
        print(' ')
    def right_view(self,start):
        print("Right view of a tree")
        while start:
            print(start.data,end=' ')
            start=start.right
root=Node(1)
root.left=Node(2)
root.right=Node(3)
root.left.left=Node(4)
root.left.right=Node(5)
root.left_view(root)
root.right_view(root)
```

## Check whether two binary tree be mirror image to each other with recursive and iterartive approach

```python
class Node:
    def __init__(self,value):
        self.value=value
        self.left=None
        self.right=None
def check(root1,root2):
    if root1 is None and root2 is None:
        return True
    if root1 is None or root2 is None:
        return False
    if root1.value != root2.value:
        return False
    else:
        return(check(root1.left,root2.right) and check(root1.right,root2.left))

def iterrative_check(root1,root2):
    sub_tree1=[]
    sub_tree2=[]
    while True:
        while root1 and root2:
            if(root1.value != root2.value):
                return False
            sub_tree1.append(root1)
            sub_tree2.append(root2)
            root1=root1.left
            root2=root2.right
        if(root1 != None and root2 != None):
            return False
        if(len(sub_tree1) != 0 and len(sub_tree2) != 0):
            root1=sub_tree1.pop()
            root2=sub_tree2.pop()
            root1=root1.right
            root2=root2.left
        else:
            break
    return True
# Creation of first tree
tree1=Node(1)
tree1.left=Node(2)
tree1.right=Node(3)
tree1.left.left=Node(4)

# Creation of second tree
tree2=Node(1)
tree2.left=Node(3)
tree2.right=Node(2)
tree2.right.right=Node(4)

if iterrative_check(tree1,tree2):
    print("Both tree is symmetric ")
else:
    print("Both tree is not symmetric")
```
