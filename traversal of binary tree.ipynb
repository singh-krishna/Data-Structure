## Binary Tree with different Traversal approach

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left,self.right=None,None
def inorder(root):
    if root is None:
        return
    if root:
        inorder(root.left)
        print(root.data,end=" ")
        inorder(root.right)
    return
def iterartive_inorder(root):
    if root is None:
        return
    s=[]
    while True:
        if root:
            s.append(root)
            root=root.left
        elif s:
            node=s.pop()
            print(node.data,end=" ")
            root=node.right
        else:
            break
def preorder(root):
    if root is None:
        return
    if root:
        print(root.data,end=" ")
        preorder(root.left)
        preorder(root.right)
    return
def iterative_preorder(root):
    s=[]
    s.append(root)
    while s:
        node=s.pop()
        print(node.data,end=" ")
        if node.right:
            s.append(node.right)
            root=node.right
        if node.left:
            s.append(node.left)
            root=node.left
def postorder(root):
    if root is None:
        return
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data,end=" ")
def iterative_postorder(root):
    if root is None:
        return
    s=[]
    s1=''
    s.append(root)
    while s:
        node=s.pop()
        s1+=str(node.data)+' '
        if node.left:
            s.append(node.left)
            root=node.left
        if node.right:
            s.append(node.right)
            root=node.right
    print(s1[::-1])
def level_order_traversal(root):
    if root is None:
        return
    s=[]
    s.append(root)
    while s:
        node=s.pop(0)
        print(node.data,end=" ")
        if node.left:
            s.append(node.left)
            root=node.left
        if node.right:
            s.append(node.right)
            root=node.right
def reversed_level_order_traversal(root):
    if root is None:
        return
    s=[]
    s1=[]
    s.append(root)
    while s:
        node=s.pop()
        s1.insert(0,node.data)
        if node.left:
            s.append(node.left)
            root=node.left
        if node.right:
            s.append(node.right)
            root=node.right
    print(s1)
        
def left_view(root):
    if root is None:
        return
    while root:
        print(root.data,end=" ")
        root=root.left
def right_view(root):
    if root is None:
        return
    while root:
        print(root.data,end=' ')
        root=root.right
def zig_zag(root):
    if root is None:
        return
    current=[]
    next=[]
    ltr=True
    current.append(root)
    while current:
        node=current.pop()
        print(node.data,end=' ')
        if ltr:
            if node.left:
                next.append(node.left)
                root=node.left
            if node.right:
                next.append(node.right)
                root=node.right
        else:
            if node.right:
                next.append(node.right)
                root=node.right
            if node.left:
                next.append(node.left)
                root=node.left
        if not current:
            ltr=not ltr
            current,next=next,current
```

```python
tree=Node(1)
tree.left=Node(2)
tree.right=Node(3)
tree.left.right=Node(5)
tree.left.left=Node(4)
print('\n Inorder')
inorder(tree)
print(' ')
iterartive_inorder(tree)
print('\n Preorder')
preorder(tree)
print(' ')
iterative_preorder(tree)
print('\nPostorder')
postorder(tree)
print(' ')
iterative_postorder(tree)
print('\n Level order traversal')
level_order_traversal(tree)
print('\n Reversed level order traversal')
reversed_level_order_traversal(tree)
print('\n Left view of tree')
left_view(tree)
print('\n Right view of tree')
right_view(tree)
print('\n Zig Zag traversal')
zig_zag(tree)
```
